```python
import numpy as np
import random as rand

dict = {}
height_list = []
for j, height in enumerate(np.linspace(1,10,10)):
    #height_list.append(height)
    #print(f"height at j={j} is {height}")
    flux_list = []
    phase_list = []
    radii_list = []
    for i,radius in enumerate(np.linspace(20,30,10)):
        #print(f'rad = {radius}')
        flux = random.uniform(0,10)
        phase = random.uniform(0,10)

        flux_list.append(flux)
        phase_list.append(phase)
        radii_list.append(radius)
    
        
    dict[j] = {
        'height': height,
        'flux': flux_list,
        'phase': phase_list,
        'radii': radii_list
    }
    height_list.append(height)
```
